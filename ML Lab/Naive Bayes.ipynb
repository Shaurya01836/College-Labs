```python
import pandas as pd 
from sklearn import tree 
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 

data = pd.read_csv("DecisionTree.csv")

data.head()

x = data.iloc[:,:-1]
x.head()

y = data.iloc[:,-1]
y.head()

a_outlook = LabelEncoder()
x.Outlook = a_outlook.fit_transform(x.Outlook)

a_temperature = LabelEncoder()
x.Temp = a_temperature.fit_transform(x.Temp)

a_Humidity = LabelEncoder()
x.Humidity = a_Humidity.fit_transform(x.Humidity)

a_Wind = LabelEncoder()
x.Wind = a_Wind.fit_transform(x.Wind)

x.head()

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.40)

classifier = GaussianNB()
classifier.fit(x_train, y_train)


print("Accuracy is : ", accuracy_score(y_test, classifier.predict(x_test)))
```
